# Estruturas de dados -  Python

# Dicionário

Um dicionário trabalha com o conceito de chave:valor e podem conter outros dicionários/listas

```python
#Declarando dicionário em branco
dic = {} 
# Verificando seu tipo
type(dic)
```

```python
#Exemplo de dicionário
pessoas = {'Gabriela':25, 'Thaís':21, 'pesos':{'Gabriela':65, 'Thais':53}, 'valores':[2,4,6,8,10], 'strings':['a','bcd']}
pessoas
```

### Acessando valores de um dicionário

```python
print(pessoas['Gabriela']) #Imprime o valor contido na chave Gabriela
print(pessoas['pesos']['Thais']) #Imprime valor contido no dicionário aninhado
print(pessoas['valores'][1]) #Imprime valor da lista contida no dicionário
print(pessoas['strings'][1]) #Imprime valor da lista contida no dicionário
```

### Manipular dicionários

```python
pessoas['Gabriela'] = 25.8 #Atualizar valor chave
pessoas['strings'][1] = 'bc' #Atualizar valor lista
del pessoas['strings'] #Remover a lista strings do dicionário

print(pessoas)
```

### Métodos de manipulação dos dicionários

```python
print(pessoas.keys()) #Imprime todas as chaves do dicionário
print(pessoas.values()) #Imprime todos os valores de todas as chaves do dicionário
print(pessoas.get('Marcos','Não existe')) #Busca pela chave, se não achar a chave, retorna Não existe
print(pessoas.setdefault('Ana',12)) #Busca pela chave, se não achar a chave, insere no dicionário
pessoas.pop('Marcos','Chave já foi removida!') #Busca pela chave e tenta remover, se não achar a chave, retorna que já foi removida
pessoas.clear() #Apaga todo o conteúdo do dicionário
print(pessoas)
```

### Atualizar um dicionário adicionando dados de outro

```python
novoDic = {'n0':2,'n1':3,'n2':4,'n5':5,'n6':'teste'}
pessoas.update(novoDic) # Atualizando dicionário pessoas com os dados do novoDic
print(pessoas)
```
