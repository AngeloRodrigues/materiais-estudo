## Ler arquivo csv

```python
file = 'databases/kc_house_data.csv' #Declarando o caminho do arquivo a ser usado nos testes
dataset = pd.read_csv(file, sep=',', header=0)
```

```python
# Plota em um gŕafico de barras o preço dos imóveis
%matplotlib notebook
dataset['price'].plot()
```

```python
# Plota em um gŕafico de barras o preço dos imóveis
%matplotlib inline
dataset['price'].plot()
```

```python
# Plotar em gráfico de dispersão o preço e o numero de quartos de imóveis
dataset.plot(x='bedrooms',y='price',kind='scatter', title='Bedrooms x Price',color='r')
```

```python
# Plotar em gráfico de dispersão o preço e o numero de banheiro de imóveis
dataset.plot(x='bathrooms',y='price',kind='scatter', title='Bathrooms x Price',color='y')
```

```python
# Plota um Histograma(distribuição de valores) com os preços na cor vermelho (bins = agrupamento de dados/ quantidade de barras vermelhas - 50 por grupo)
dataset['price'].hist(bins=50, color='red') # Maior quantidade de imóveis está de 0 a 1000000
```

```python
# Plota histogramas das colunas bedrooms e bathrooms lado a lado na cor verde.
dataset[['bedrooms','bathrooms']].hist(bins=30,alpha=0.5,color='Green') #alpha = opacidade da cor (verde mais opaco)
```
